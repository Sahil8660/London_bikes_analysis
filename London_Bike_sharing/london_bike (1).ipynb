```python
new_col_dict = {'timestamp':'time',
                'cnt':'count',
                't1':'temp_real_C',
                't2':'temp_feels_like_C',
                'hum':'humidity_per',
                'wind_speed':'wind_speed_percent',
                'weather_code':'weather',
                'is_holiday':'is_holiday',
                'is_weekend':'is_weekend',
                'season':'season'
                }
bikes.rename(new_col_dict,axis=1,inplace=True)
```

```python
bikes.humidity_per = bikes.humidity_per/100
```

```python
bikes.head()
```

```python
bikes.season.value_counts().sort_values(ascending=True)
```

```python
#changing season number to names of the season
season_dic = {
    '0.0':'Spring',
    '1.0':'Summer',
    '2.0':'Autumn',
    '3.0':'Winter'}
```

```python
bikes.weather.value_counts()
```

```python
#changing weather number to the type of weather condition
weather_dic ={
    '1.0':'Clear',
    '2.0':'Scattered Clouds',
    '3.0':'Broken Clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with Thunderstorms',
    '26.0':'Snowfall'
}
```

```python
bikes.season=bikes.season.astype(str)
bikes.season = bikes.season.map(season_dic)
```

```python
bikes.weather = bikes.weather.astype(str)
bikes.weather = bikes.weather.map(weather_dic)
```
